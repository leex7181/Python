```python
# Cross join
#1. All rows from Table "a"' will be mached" will all rows from Table "b. 2. We do not need a "key" column to Join On 3. There will be Duplic- Be careful how vou use Cross Joins

# No pre-existing funtion I am aware for Cross Join - Hence we need to create a Function manually

def cross_join(table1,table2):
    return(table1.assign(key=1).merge(table2.assign(key=1), on='key').drop('key',1))
```
